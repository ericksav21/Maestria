```python
from matplotlib import pyplot as plt
from math import *
from numpy import linspace
```

```python

omega = pi + 0.5
A = 1
phi = 0
dominio = [i for i in range(11)]

print(dominio)

k = 0

omegaP = omega
while omegaP > 2*pi:
    omegaP -= 2*pi

    omegaP -= 2*pi
    
print(omegaP)
    
f1 = lambda x: A*cos(omega*x + phi)
f2 = lambda x: A*cos(omegaP*x + phi)

cod1 = list(map(f1, dominio))
cod2 = list(map(f2, dominio))

for i in range(len(cod1)):
    
    print(cod1[i], cod2[i])

C = linspace(0, 10, 1000)

plt.plot(C, list(map(f1, C)))
plt.plot(C, list(map(f2, C)))

plt.scatter(dominio, cod1)

plt.show()
    
```
