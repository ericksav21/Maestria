## Demostrar que para frecuencias $|\omega| > \pi$ estamos generando una señal "Alias" con una frecuencia $|\omega| < \pi$

Primero hay que demostrar que una señal discreta es periódica sí y solo sí la frecuencia es un número racional. Es decir: $x(n + N) = x(n)$ $\forall n > 0 \land N = \frac{p}{q}$.

Para demostrar esto usamos una sinusoide de frecuencia $f_0$: $x(n) = cos(2 \pi f_0 n + \theta)$.

Ahora, usando la periodicidad del conseno tenemos:

$$cos(2 \pi f_0(N + n) + \theta) = cos(2 \pi f_0 n + \theta)$$

Para que lo anterior sea cierto se debe cumplir que:

$$2 \pi f_0 N = 2k \pi$$ ó $$f_0 = \frac{k}{N}$$ donde $k, N \in Z$

Ahora, se mostrará que las señales discretas en el tiempo con frecuencias separadas por múltiplos de $2 \pi$ son idénticas.

Tenemos una señal sinusoidal discreta:

$$s(n) = A \cos(\omega n + \theta)\ \forall n \in Z$$

Como $n$ es entero y la función coseno es periódica con periodo $2 \pi$ tenemos:

$$s(n) = A \cos(n(\omega + 2k \pi) + \theta)$$
$$= A \cos(n \omega + 2kn \pi + \theta)$$
$$= A \cos(n \omega + \theta)$$

Esto anterior es válido porque tanto $k$ y $n$ son enteros.
Por lo tanto las señales sinusolidales con frecuencia $w_k = w_0 + 2 \pi k$ son idénticas. Las señales con frecuencia $|\omega| < \pi$ son las señales espectro y las que están fuera de dicho rango son las conocidas como $\textit{Alias}$.

Lo único que falta es demostrar que para alguna señal sinusoidal, la frecuencia $\omega$ está dentro del rango $(-\pi, \pi)$. Para ello podemos tomar un $\omega_k > \pi$ y lo usamos en la señal propuesta anteriormente.

$$s(n) = A \cos(n(\omega_k + 2 \pi k) + \theta)$$

Como $\omega_k > \pi$ podemos hacer una descomposición de tal forma que nos quede un múltiplo de $\pi$ más un residuo:

$$\omega_k = m_k \pi + r$$ Donde: $m_k \in Z^+$ y $r \in R$ y $0 < r < \pi$

Sustituyendo $\omega_k$ en la señal nos queda:

$$s(n) = A \cos(n(m_k \pi + r + 2 \pi k) + \theta)$$

Como $m_k$ es entero entonces podemos factorizar $m_k \pi$ de tal forma que nos quede:

$$s(n) = A \cos(n(r + [2k + m_k + r ] \pi) + \theta)$$
$$s(n) = A \cos(nr + n[2k + m_k + r ] \pi + \theta)$$

Para poder aplicar la periodicidad del cosena en la ecuación anterior tenemos que ver la paridad de $m_k$ por lo cual se nos generan dos casos:

1)$m_k$ es par:

En este caso el término $n[2k + m_k]$ es par y por lo tanto: $A \cos(nr + n[2k + m_k] \pi + \theta) = A \cos(nr + \theta)$. Se cumple el Alias.

2)$m_k$ es impar:

Para este caso podemos sumar y restar $\pi$ a la definición de $\omega_k$:

$$\omega_k = m_k \pi + r + \pi - \pi$$
$$= 2 m_k \pi + (r - \pi)$$

De lo anterior podemos ver que de nuevo tenemos un múltiplo de $2 \pi$ aunque tenemos que ver lo que pasa con el término $r - \pi$. Lo primero que podemos destacar es que está en el intervalo $(-\pi, 0)$ por lo que nos queda:

$$A \cos(n(r - \pi) + 2n \pi [k + m_k] + \theta) = A \cos(n (r - \pi) + \theta)$$

De esta forma se demuestra que $\omega < \pi$, haciendo un análisis similar se puede ver que $-\pi < \omega$.

Todo esto también se puede demostrar usando la función seno dado que tiene la misma periodicidad que el coseno.

## Generar una señal con una frecuencia mayor a $\pi$ y generar la señal alias. Graficar ambas gráficas y verificar que son iguales.

Primero vamos a calcular unas señales con periodo $\frac{k}{256}$ y calcularemos algunos alias.

```python
from matplotlib import pyplot as plt
from math import *
import numpy as np

def alias(w, a, t, theta):
    alias1 = w + 2 * np.pi
    alias2 = w + 4 * np.pi
    
    x = np.arange(0, 20, t, dtype = 'int')
    y0 = a * np.cos(w * x + theta)
    y1 = a * np.cos(alias1 * x + theta)
    y2 = a * np.cos(alias2 * x + theta)
    
    return x, y0, y1, y2

w1 = 128.0 / 256.0

%matplotlib inline
x, y0, y1, y2 = alias(w1, 3, 1, 0)

plt.plot(x, y0)
plt.title("Muestreo de una señal con w = " + str(w1))
```

```python
plt.plot(x, y1)
plt.title("Muestreo de una señal con w = 2 * pi * w")
```

```python
plt.plot(x, y2)
plt.title("Muestreo de una señal con w = 4 * pi * w")
```

```python
w2 = 172.0 / 256.0
x, y0, y1, y2 = alias(w2, 5, 1, 0)

plt.plot(x, y0)
plt.title("Muestreo de una señal con w = " + str(w2))
```

```python
plt.plot(x, y1)
plt.title("Muestreo de una señal con w = 2 * pi * w")
```

```python
plt.plot(x, y2)
plt.title("Muestreo de una señal con w = 4 * pi * w")
```

```python
w3 = 200.0 / 256.0
x, y0, y1, y2 = alias(w3, 2, 1, 0)

plt.plot(x, y0)
plt.title("Muestreo de una señal con w = " + str(w3))
```

```python
plt.plot(x, y1)
plt.title("Muestreo de una señal con w = 2 * pi * w")
```

```python
plt.plot(x, y2)
plt.title("Muestreo de una señal con w = 4 * pi * w")
```

Podemos ver en las gráficas anteriores que al usar alias de una señal con $\omega < \pi$ las muestras son iguales.

## Calcular cual es la frecuencia alias de una señal con frecuencia < $\pi$

```python
omega = 2 * pi + 0.5
a = 1
phi = 0
x = np.arange(0, 11, dtype = 'int')

#Calcular el espectro
omegaP = omega
while omegaP > 2*pi:
    omegaP -= 2*pi

#omegaP -= 2*pi

print("El espectro de la señal es: " + str(omegaP))
    
f1 = lambda x: a * np.cos(omega * x + phi)
f2 = lambda x: a * np.cos(omegaP * x + phi)

cod1 = list(map(f1, x))
cod2 = list(map(f2, x))

for i in range(len(cod1)):
    print(cod1[i], cod2[i])

C = np.linspace(0, 10, 1000)

plt.plot(C, list(map(f1, C)), color = "blue")
plt.plot(C, list(map(f2, C)), color = "orange")

plt.scatter(x, cod1)

plt.show()
```

Podemos ver en la gráfica anterior una señal (naranja) con su alias (azul).
De la misma forma podemos verificar que las dos señales son iguales si las muestreamos con un mismo tamaño de paso.

```python
f1 = a * np.cos(omega * x + phi)

plt.plot(x, f1)
plt.title("Señal con frecuencia w = " + str(omega) + " (Alias)")
```

```python
f2 = a * np.cos(omegaP * x + phi)

plt.plot(x, f2)
plt.title("Señal con frecuencia w = " + str(omegaP))
```
